```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
root = "result.csv"
df = pd.read_csv(root)
df
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Data Size"], df["Insertion Sort"], marker="o", label="Insertion Sort")
plt.plot(df["Data Size"], df["Selection Sort"], marker="o", label="Selection Sort")
plt.plot(df["Data Size"], df["Bubble Sort"], marker="o", label="Bubble Sort")

plt.title("Sort Algorithm Avg")
plt.xlabel("(Data Size)")
plt.ylabel("(s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
